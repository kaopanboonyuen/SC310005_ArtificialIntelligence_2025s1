# ‚úÖ Step 2: Import libraries

```python
import os, zipfile, glob, random, cv2, torch
import matplotlib.pyplot as plt
from ultralytics import YOLO
from IPython.display import Image, display
```

# üì• Step 3: Download Brain Tumor Dataset (from Kao's repo)

```python
!wget -O brain-tumor.zip https://github.com/kaopanboonyuen/panboonyuen_dataset/raw/main/public_dataset/ultralytics_dataset/YOUR_DATASET.zip
```

# ‚úÖ Unzip dataset

```python
dataset_dir = "/content/datasets/YOUR_DATASET"
os.makedirs(dataset_dir, exist_ok=True)
with zipfile.ZipFile("brain-tumor.zip", 'r') as zip_ref:
    zip_ref.extractall("/content/datasets/YOUR_DATASET")

print("‚úÖ Dataset extracted to:", dataset_dir)
print("üìÇ Files inside:", os.listdir(dataset_dir))
```

# üìù Step 4: Rewrite brain-tumor.yaml for Colab path

```python
yaml_content = """# Brain Tumor Detection Dataset (Ultralytics format)
path: /content/datasets/brain-tumor

train: images/train
val: images/val
test:

names:
  0: YOUR_CLASS_NAME
  1: YOUR_CLASS_NAME

download: https://github.com/kaopanboonyuen/panboonyuen_dataset/raw/main/public_dataset/ultralytics_dataset/brain-tumor.zip
"""
with open(os.path.join(dataset_dir, "brain-tumor.yaml"), "w") as f:
    f.write(yaml_content)

print("‚úÖ Rewritten brain-tumor.yaml")
!cat /content/datasets/brain-tumor/brain-tumor.yaml
```

# üëÄ Step 5: Preview dataset images

```python
train_images = glob.glob(os.path.join(dataset_dir, "images/train/*.jpg"))
sample_images = random.sample(train_images, 4)

plt.figure(figsize=(12, 8))
for i, img_path in enumerate(sample_images):
    img = cv2.imread(img_path)[..., ::-1]  # BGR -> RGB
    plt.subplot(2, 2, i+1)
    plt.imshow(img)
    plt.axis("off")
    plt.title(f"Train Sample {i+1}")
plt.show()
```

# üöÄ Step 6: Train YOLOv8 on Brain Tumor Dataset

```python
model = YOLO("YOUR_PRETRAINED_MODEL")  # small & fast model

results = model.train(
    data="/content/datasets/brain-tumor/YOUR_YAML_FILENAME.yaml",
    epochs=#INSERT_YOUR_FAV_EPOCH_NUMBERS,          # try 50-100 for better results
    imgsz=640,
    batch=16,
    name="yolo_brain_tumor"
)
```

# üìä Step 7: Visualize Training Results (Loss & mAP curves)

```python
plt.figure(figsize=(12,6))
results_path = os.path.join(model.trainer.save_dir, "results.png")
if os.path.exists(results_path):
    display(Image(filename=results_path))
else:
    print("‚ö†Ô∏è No results.png found yet.")
```

# üìà Step 8: Evaluate on Validation Set + Extra Plots

```python
metrics = model.val(save=True, plots=True)  # plots=True ‡∏à‡∏∞ gen confusion matrix, PR curve
print("‚úÖ Evaluation metrics:", metrics)

# Display generated metrics plots
eval_dir = model.trainer.save_dir
for plot_name in ["confusion_matrix.png", "PR_curve.png", "F1_curve.png", "P_curve.png", "R_curve.png"]:
    plot_path = os.path.join(eval_dir, plot_name)
    if os.path.exists(plot_path):
        print(f"üìä Showing {plot_name}")
        display(Image(filename=plot_path))
```

# üîÆ Step 9: Inference on a Random Validation Image

```python
from pathlib import Path
```

```python
val_images = glob.glob(os.path.join(dataset_dir, "images/val/*.jpg"))
test_img = random.choice(val_images)

results = model(test_img, save=True)
print("üîç Prediction done on:", test_img)

pred_img = Path(results[0].save_dir) / Path(results[0].path).name

display(Image(filename=pred_img))
```

### üîπ Section 2: Data Structures

#### ‚úÖ Concepts:

* List, Tuple, Dictionary, Set
* Indexing and slicing
* List comprehension

```python
# List
fruits = ["apple", "banana", "mango"]
print(fruits[0])  # indexing
print(fruits[1:]) # slicing

# Dictionary
student = {"name": "John", "score": 95}
print(student["name"])

# Set
numbers = {1, 2, 3, 2}
print(numbers)  # no duplicates

# List comprehension
squares = [x**2 for x in range(5)]
print(squares)
```

### üîπ Section 6: Plotting with Matplotlib

#### ‚úÖ Concepts:

* Simple plotting with `matplotlib.pyplot`

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [i**2 for i in x]

plt.plot(x, y)
plt.title("Square Function")
plt.xlabel("x")
plt.ylabel("y = x^2")
plt.grid(True)
plt.show()
```

# ‚úÖ Step 2: Import libraries

```python
import os, zipfile, glob, random, cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO
from IPython.display import Image, display
```

# üì• Step 3: Download Crack Segmentation Dataset

```python
!wget -O crack-seg.zip https://github.com/kaopanboonyuen/panboonyuen_dataset/raw/main/public_dataset/ultralytics_dataset/YOUR_DATASET.zip

dataset_dir = "/content/datasets/YOUR_DATASET"
os.makedirs(dataset_dir, exist_ok=True)

with zipfile.ZipFile("crack-seg.zip", 'r') as zip_ref:
    zip_ref.extractall("/content/datasets/YOUR_DATASET")

print("‚úÖ Dataset extracted to:", dataset_dir)
print("üìÇ Files inside:", os.listdir(dataset_dir))
```

# üìù Step 4: Rewrite crack-seg.yaml for Colab path

```python
yaml_content = f"""# Crack Segmentation Dataset (Ultralytics format)
path: {dataset_dir}

train: images/train
val: images/val
test: images/test

# Classes
names:
  0: YOUR_CLASS_NAME

download: https://github.com/kaopanboonyuen/panboonyuen_dataset/raw/main/public_dataset/ultralytics_dataset/crack-seg.zip
"""

with open(os.path.join(dataset_dir, "crack-seg.yaml"), "w") as f:
    f.write(yaml_content)

print("‚úÖ Rewritten crack-seg.yaml")
!cat /content/datasets/crack-seg/crack-seg.yaml
```

# üëÄ Step 5: Preview dataset images

```python
train_images = glob.glob(os.path.join(dataset_dir, "images/train/*.jpg"))
sample_images = random.sample(train_images, 4)

plt.figure(figsize=(12, 8))
for i, img_path in enumerate(sample_images):
    img = cv2.imread(img_path)[..., ::-1]
    plt.subplot(2, 2, i+1)
    plt.imshow(img)
    plt.axis("off")
    plt.title(f"Train Sample {i+1}")
plt.show()
```

# üöÄ Step 6: Train YOLOv8 Segmentation Model

```python
model = YOLO("YOUR_PRETRAINED_MODEL")  # segmentation model (nano version, fast)

results = model.train(
    data=f"{dataset_dir}/crack-seg.yaml",
    epochs=#INSERT_YOUR_FAV_EPOCH_NUMBERS,
    imgsz=640,
    batch=16,
    name="yolo_crack_seg"
)
```

# üìä Step 7: Visualize Training Results

```python
results_path = os.path.join(model.trainer.save_dir, "results.png")
if os.path.exists(results_path):
    display(Image(filename=results_path))
```

# üìà Step 8: Evaluate Model Performance

```python
metrics = model.val(save=True, plots=True)
print("‚úÖ Evaluation metrics:", metrics)

eval_dir = model.trainer.save_dir
for plot_name in ["confusion_matrix.png", "PR_curve.png", "F1_curve.png"]:
    plot_path = os.path.join(eval_dir, plot_name)
    if os.path.exists(plot_path):
        display(Image(filename=plot_path))
```

# üîÆ Step 9: Inference on Random Validation Image (Segmentation)

```python
val_images = glob.glob(os.path.join(dataset_dir, "images/val/*.jpg"))
test_img = random.choice(val_images)

results = model(test_img, save=True)
print("üîç Prediction done on:", test_img)

# ‚úÖ Fix path join
pred_dir = results[0].save_dir
pred_img = os.path.join(pred_dir, os.path.basename(results[0].path))

display(Image(filename=pred_img))
```
